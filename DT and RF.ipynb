```python
import numpy as np
import pandas as pd
from sklearn import tree

df = pd.read_csv('Loan_fic.csv')
```

```python
df.head()
```

```python
d ={'Y':1, 'N':0}
df['Offered'] = df['Offered'].map(d)
df['Employed?'] = df['Employed?'].map(d)
df['Has existing policy'] = df['Has existing policy'].map(d)
df['Public records OK'] = df['Public records OK'].map(d)
d = {'BS': 0, 'MS': 1, 'PhD': 2}
df['Level of Education'] = df['Level of Education'].map(d)
df.head(13)
```

```python
features = list(df.columns[:6])
features
```

```python
y = df['Offered']
X = df[features]
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)
```

```python
tree.export_graphviz(clf, out_file = 'loan_fic.dot', feature_names = features)
```
